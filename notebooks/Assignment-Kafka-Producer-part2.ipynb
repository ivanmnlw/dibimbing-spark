```python
import json
import uuid
import os
import json
from dotenv import load_dotenv
from pathlib import Path
from kafka import KafkaProducer
from faker import Faker
from time import sleep
from datetime import datetime, timedelta
```

```python
dotenv_path = Path('/resources/.env')
load_dotenv(dotenv_path=dotenv_path)
```

```python
kafka_host = os.getenv("KAFKA_HOST")
kafka_topic = "dibimbing_assignment_2"
```

```python
print(kafka_host)
```

```python
producer = KafkaProducer(bootstrap_servers=f'{kafka_host}:9092')
```

```python
_instance = Faker()
global faker
faker = Faker()
```

```python
class DataGenerator(object):
    @staticmethod
    def get_data():
        now = datetime.now()
        return [
            uuid.uuid4().__str__(),
            faker.uuid4(),
            faker.uuid4(),
            faker.word().capitalize(),
            faker.random_element(elements=("Electronics", "Clothing", "Home Goods", "Books", "Toys")),
            faker.random_int(min=1, max=5),
            faker.random_int(min=5, max=100),
            faker.random_element(elements=("credit_card", "debit_card", "paypal", "cash")),
            faker.unix_time(
                start_datetime=now - timedelta(minutes=60), end_datetime=now
            )
        ]

for i in range(1,400):
    columns =  ["event_id", "user_id", "item_id", "item_name", "category", "quantity", "price", "payment_method", "ts"]
    data_list = DataGenerator.get_data()
    json_data = dict(
        zip(columns,data_list)
    )
    _payload = json.dumps(json_data).encode("utf-8")
    response = producer.send(topic=kafka_topic, value=_payload)
    print(_payload)
    sleep(2)
```
